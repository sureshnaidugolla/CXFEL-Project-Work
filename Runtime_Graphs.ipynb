```python
import matplotlib.pyplot as plt
import numpy as np

str2="6.1	5.7	5.7	6.9	7.6	9.3	16.0	43.5	181.1	677.7	3697.5"
str4="9.7	10.3	9.2	10.5	12.5	25.8	70.6	297.2	1444.0	6483.5	70814.9"
str6="6.1	5.5	5.8	6.1	13.4	50.0	203.3	999.0	4488.6"
str8="6.2	7.2	7.6	13.8	31.9	112.7	507.4	2563.8	12578.7"
str10="6.4	7.5	8.2	18.4	61.8	213.8	998.4	5961.7	28997.0"

pix2=list(map(float,str2.split()))
pix4=list(map(float,str4.split()))
pix6=list(map(float,str6.split()))
pix8=list(map(float,str8.split()))
pix10=list(map(float,str10.split()))
snapshots=[1,2,4,8,16,32,64,128,256,512,1024]

plt.xlabel("Snapshots")
plt.ylabel("Time (s)")
plt.plot(snapshots[:len(pix6)],pix2[:len(pix6)],label="2 pixels")
plt.plot(snapshots[:len(pix6)],pix4[:len(pix6)],label="4 pixels")
plt.plot(snapshots[:len(pix6)],pix6,label="6 pixels")
plt.plot(snapshots[:len(pix8)],pix8,label="8 pixels")
plt.plot(snapshots[:len(pix10)],pix10,label="10 pixels")
plt.yscale('log')
plt.legend()
```
