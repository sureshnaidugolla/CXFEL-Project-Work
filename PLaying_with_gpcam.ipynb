```python
# Evaluating the synthetic test functions
##################################################################

import numpy as np
from gpcam.autonomous_experimenter import AutonomousExperimenterFvGP

#first synthetic test function
def test_function1(p0,p1):
  if p0>=30:
    return 8-(2*p0*np.sin(np.sqrt(np.abs(p0)))+2*p1*np.sin(np.sqrt(np.abs(p1))))+140
  elif(np.sqrt(p0*p0+p1*p1)>=20):
    return 8-(2*p0*np.sin(np.sqrt(np.abs(p0)))+2*p1*np.sin(np.sqrt(np.abs(p1))))+160
  else:
    return 8-(2*p0*np.sin(np.sqrt(np.abs(p0)))+2*p1*np.sin(np.sqrt(np.abs(p1))))

#second synthetic test function
def test_function2(p0,p1):
  if p0>=30:
    return 8-(2*p0*np.sin(np.sqrt(np.absolute(p0)))+2*p1*np.sin(np.sqrt(np.absolute(p1))))+200
  elif(np.sqrt(p0*p0+p1*p1)>=20):
    return 8-(2*p0*np.sin(np.sqrt(np.absolute(p0)))+2*p1*np.sin(np.sqrt(np.absolute(p1))))+360
  else:
    return 8-(2*p0*np.sin(np.sqrt(np.absolute(p0)))+2*p1*np.sin(np.sqrt(np.absolute(p1))))

def instrument1(data,instrument_dict={}):
  for entry in data:
    pix0 = test_function1(entry["x_data"][0],entry["x_data"][1])
    entry["y_data"] =np.asarray([pix0])
    entry["noise variances"] = 0.1
    entry["output positions"] = np.array([[0]])
  return data

def instrument2(data,instrument_dict={}):
  for entry in data:
    pix0 = test_function2(entry["x_data"][0],entry["x_data"][1])
    entry["y_data"] =np.asarray([pix0])
    entry["noise variances"] = 0.1
    entry["output positions"] = np.array([[0]])
  return data

parameter_bounds = np.array([[0,50],[0,50]])
my_ae1 = AutonomousExperimenterFvGP(parameter_bounds,
                                  2,
                                  1,
                                  instrument_function=instrument1,
                                  init_dataset_size=6
                                  )

my_ae2 = AutonomousExperimenterFvGP(parameter_bounds,
                                  2,
                                  1,
                                  instrument_function=instrument1,
                                  init_dataset_size=6
                                  )
```

```python
import numpy as np
def test_function1(p0,p1):
  if p0>=30:
    return 8-(2*p0*np.sin(np.sqrt(np.absolute(p0)))+2*p1*np.sin(np.sqrt(np.absolute(p1))))+140
  elif(np.sqrt(p0*p0+p1*p1)>=20):
    return 8-(2*p0*np.sin(np.sqrt(np.absolute(p0)))+2*p1*np.sin(np.sqrt(np.absolute(p1))))+160
  else:
    return 8-(2*p0*np.sin(np.sqrt(np.absolute(p0)))+2*p1*np.sin(np.sqrt(np.absolute(p1))))

def test_function2(p0,p1):
  if p0>=30:
    return 8-(2*p0*np.sin(np.sqrt(np.absolute(p0)))+2*p1*np.sin(np.sqrt(np.absolute(p1))))+200
  elif(np.sqrt(p0*p0+p1*p1)>=20):
    return 8-(2*p0*np.sin(np.sqrt(np.absolute(p0)))+2*p1*np.sin(np.sqrt(np.absolute(p1))))+360
  else:
    return 8-(2*p0*np.sin(np.sqrt(np.absolute(p0)))+2*p1*np.sin(np.sqrt(np.absolute(p1))))

x1=np.linspace(0,50,100)
x2=np.linspace(0,50,100)
X1,X2=np.meshgrid(x1,x2)

Z1 = np.zeros_like(X1)
Z2 = np.zeros_like(X1)

# Calculate the Z values using the test_function
for i in range(X1.shape[0]):
    for j in range(X1.shape[1]):
        Z1[i, j] = test_function1(X1[i, j], X2[i, j])
        Z2[i, j] = test_function2(X1[i, j], X2[i, j])
```

```python
import plotly.graph_objects as go

fig1 = go.Figure(data=[go.Surface(x=X1,y=X2,z=Z1)])
fig1.update_layout(title='Test Function plot(a)', autosize=True)
fig1.show()

fig2 = go.Figure(data=[go.Surface(x=X1,y=X2,z=Z2)])
fig2.update_layout(title='Test Function plot(b)', autosize=True)
fig2.show()
```

```python
import numpy as np
from scipy.optimize import minimize

# Define the objective function
def objective_function(x):
    return np.sum(x**2)

# Initial guess for the parameters
x0 = np.array([1, 1])

# Perform optimization using the BFGS algorithm
result = minimize(objective_function, x0, method='BFGS', tol=1e-6)

# Print the result
print(result)
```
