```python
import numpy as np
import statistics

def new_dataFilter(y):
  diff = np.diff(y)
  print((statistics.mode(diff)+statistics.median(diff) )/2)
  for i in range(1,len(y)):
    y[i]=y[i-1]+(statistics.mode(diff)+statistics.median(diff) )/2
  print(y)
  return y

```

```python
y=np.asarray([1,2,3,10,15,20,120,130,140])
from scipy.interpolate import make_interp_spline
from scipy.signal import medfilt
import matplotlib.pyplot as plt
plt.plot(y)
plt.plot(new_dataFilter(y))
plt.plot(make_interp_spline(np.arange(len(y)),y)(np.linspace(min(y), max(y), len(y))))
plt.plot(medfilt(y, kernel_size=len(y)))
plt.plot(y[np.abs(y-np.mean(y))<np.std(y)])
plt.legend(['original', 'filtered','interpolated','scipy medfilt','using standard deviation'])
plt.show()
```
